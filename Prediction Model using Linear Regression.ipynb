```python
import pandas as pd
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt
```

```python
url ='https://raw.githubusercontent.com/AdelOuledSaid/Machine-learning-with-Apache-Spark-/main/Prediction%20Model%20using%20Linear%20Regression/mpg%20(1).csv'
df= pd.read_csv(url)
```

```python
df.sample(5)
```

```python
df.shape
```

```python
df.plot.scatter( x="Horsepower" , y = "MPG")
```

#### Identify the target column and the data columns

```python
target = df["MPG"]
```

Identify the features columns 

```python
features = df[['Horsepower', 'Weight']]
```

 Build and Train a Linear Regression Model


```python
lr = LinearRegression() #create a model 
```

```python
lr.fit(features , target )  # FIT MODEL
```

#### Evaluate the model and make predictions

```python
lr.score(features , target)
```

```python
predict = lr.predict([[100,2000]])
```
